
# Creating **IMG-IMU pairs** within a computed window
*   Loads synchronized data from extracted P000 folder (images + IMU + timestamps)
* Uses a custom get_imu_window function with 200ms window, 20 samples, 10 min threshold
* Creates image-IMU pairs with proper time synchronization
* Skips frames that don't meet minimum IMU sample requirements
* Prepares batches in exact format for our model: (batch_size, 3, 224, 224) for images and (batch_size, 6, 20) for IMU
* Saves synchronized pairs to pickle file for training





```python
import os
import numpy as np
import pickle
from PIL import Image

root_path = "/content/drive/MyDrive/tartanair_v2"
env = "Gascola"
traj = "easy"
file_directory = '/content/drive/MyDrive/tartanair_v2/unziped_imu/Gascola/Data_easy/P000/imu'
# # Path to the extracted data
# img_p000_path = os.path.join("unziped_image_lcam_front",env, f"Data_{traj}", "P000")
# imu_p000_path = os.path.join("unziped_imu","Gascola", "Data_easy", "P000")
imu_p000_path = '/content/drive/MyDrive/tartanair_v2/unziped_imu/Gascola/Data_easy/P000'
img_p000_path = '/content/drive/MyDrive/tartanair_v2/unziped_image_lcam_front/Gascola/Data_easy/P000'

# Load timestamps and IMU data
cam_time= np.load(os.path.join(file_directory, 'cam_time.npy'))
imu_time= np.load(os.path.join(file_directory, 'imu_time.npy'))
print(cam_time, 'debug tiiiime')
acc = np.load(os.path.join(imu_p000_path, "imu", "acc_nograv_body.npy"))
gyro = np.load(os.path.join(imu_p000_path, "imu", "gyro.npy"))
imu = np.concatenate([acc, gyro], axis=1)  # shape: (N_imu, 6)

# Get sorted image filenames
img_folder = os.path.join(img_p000_path, "image_lcam_front")
img_files = sorted([f for f in os.listdir(img_folder) if f.endswith(".png")])

print(f"Images: {len(img_files)}, Cam timestamps: {len(cam_time)}, IMU: {imu.shape}")

# Your custom function (verified accurate!)
def get_imu_window(imu_time, imu_data, t_img, window_duration=0.2,
                   target_samples=20, min_samples=10, forward_only=False):
    """
    Extract fixed-length IMU window around image timestamp.

    Args:
        imu_time (np.array): shape (N,), IMU timestamps
        imu_data (np.array): shape (N, C), IMU measurements
        t_img (float): timestamp of image frame
        window_duration (float): total window length in seconds
        target_samples (int): desired number of samples
        min_samples (int): minimum samples threshold
        forward_only (bool): if True, window = [t_img, t_img+window_duration],
                             else centered around t_img

    Returns:
        np.array shape (target_samples, C) or None if skip
    """
    if forward_only:
        start = t_img
        end    = t_img + window_duration
    else:
        start = t_img - window_duration / 2
        end    = t_img + window_duration / 2

    mask = (imu_time >= start) & (imu_time <= end)
    segment = imu_data[mask]

    if len(segment) < min_samples:
        print(f"Skipping frame at t={t_img:.3f}s: only {len(segment)} IMU samples (<{min_samples})")
        return None

    # If too many samples, down-sample uniformly
    if len(segment) > target_samples:
        idx = np.linspace(0, len(segment)-1, target_samples).astype(int)
        segment = segment[idx]

    # If too few but above min_samples, pad by repeating last sample
    if len(segment) < target_samples:
        pad_count = target_samples - len(segment)
        pad = np.repeat(segment[-1:,:], pad_count, axis=0)
        segment = np.vstack([segment, pad])

    return segment

# Create synchronized pairs with your custom function
pairs = []
skipped_count = 0

for i in range(min(len(img_files), len(cam_time))):
    t_img = cam_time[i]
    img_name = img_files[i]

    # Get corresponding IMU window using your function
    imu_win = get_imu_window(imu_time, imu, t_img,
                            window_duration=0.2,      # 200ms window
                            target_samples=20,        # 20 samples
                            min_samples=10,           # minimum 10 samples
                            forward_only=False)       # centered around image

    if imu_win is not None:
        pairs.append({
            'image_path': os.path.join(img_folder, img_name),
            'image_timestamp': t_img,
            'imu_data': imu_win  # shape: (20, 6)
        })
    else:
        skipped_count += 1

print(f"✅ Created {len(pairs)} synchronized image-IMU pairs")
print(f"⚠️  Skipped {skipped_count} frames due to insufficient IMU data")

# Function to prepare batch data for model training
def prepare_batch_data(pairs, batch_size=32, img_size=(224, 224)):
    """
    Prepare batch data for V-ICE model training

    Args:
        pairs: list of synchronized image-IMU pairs
        batch_size: number of samples per batch
        img_size: target image size (H, W)

    Returns:
        images: numpy array of shape (batch_size, 3, H, W)
        imu_ numpy array of shape (batch_size, 6, seq_len)
    """
    if len(pairs) == 0:
        raise ValueError("No valid pairs available for batching")

    # Randomly select batch indices
    batch_indices = np.random.choice(len(pairs), min(batch_size, len(pairs)), replace=False)

    images = []
    imu_data_list = []

    for idx in batch_indices:
        pair = pairs[idx]

        # Load and preprocess image
        img = Image.open(pair['image_path']).convert('RGB')
        img = img.resize(img_size)
        img_array = np.array(img).transpose(2, 0, 1).astype(np.float32) / 255.0  # Normalize to [0,1], shape: (3, H, W)
        images.append(img_array)

        # Get IMU data and transpose for model input
        imu_raw = pair['imu_data']  # shape: (seq_len, 6) - now (20, 6)
        imu_tensor = imu_raw.T  # shape: (6, seq_len) - now (6, 20) for model input
        imu_data_list.append(imu_tensor)

    images = np.stack(images)  # shape: (batch_size, 3, H, W)
    imu_data = np.stack(imu_data_list)  # shape: (batch_size, 6, seq_len)

    return images, imu_data

# Test with a small batch
if pairs:
    batch_images, batch_imu = prepare_batch_data(pairs, batch_size=8)

    print(f"\n✅ Perfect input shapes for V-ICE model:")
    print(f"Image Batch: {batch_images.shape}  <- Expected: (batch_size, 3, 224, 224)")
    print(f"IMU Batch: {batch_imu.shape}     <- Expected: (batch_size, 6, 20) with your function")

    # Verify data ranges
    print(f"\nData verification:")
    print(f"Image range: [{batch_images.min():.3f}, {batch_images.max():.3f}]  <- Should be [0, 1]")
    print(f"IMU range: [{batch_imu.min():.3f}, {batch_imu.max():.3f}]  <- Should be reasonable values")

# Save pairs for later use
output_dir = os.path.join(root_path, env, f"Data_{traj}")
os.makedirs(output_dir, exist_ok=True)
with open(os.path.join(output_dir, "pairs_step1.pkl"), "wb") as f:
    pickle.dump(pairs, f)

print(f"\n✅ All {len(pairs)} pairs saved to: {os.path.join(output_dir, 'pairs_step1.pkl')}")
print("✅ Data is ready for V-ICE training!")
```

```python
import os
import numpy as np
import pickle
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import time

# Your model code - FIXED
class IMUEncoder(nn.Module):
    def __init__(self, input_channels=6, seq_len=20, latent_dim=128):
        super(IMUEncoder, self).__init__()
        self.conv1 = nn.Conv1d(input_channels, 32, kernel_size=3, padding=1)
        self.conv2 = nn.Conv1d(32, 64, kernel_size=3, padding=1)
        self.pool = nn.AdaptiveAvgPool1d(1)
        self.fc1 = nn.Linear(64, latent_dim)
        self.fc2 = nn.Linear(latent_dim, latent_dim)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.1)

    def forward(self, x):
        x = self.relu(self.conv1(x))
        x = self.dropout(x)
        x = self.relu(self.conv2(x))
        x = self.dropout(x)
        x = self.pool(x)
        x = x.view(x.size(0), -1)  # Shape: (batch_size, 64)
        x = self.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x

class VisionEncoderWithProjection(nn.Module):
    def __init__(self, base_model, latent_dim=128):
        super(VisionEncoderWithProjection, self).__init__()
        self.base = base_model
        # Fix: Get the correct feature dimension from the vision model
        base_output_dim = getattr(base_model, 'num_features', 2048)  # Default to 2048
        self.projection_head = nn.Sequential(
            nn.Linear(base_output_dim, latent_dim),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(latent_dim, latent_dim)
        )

    def forward(self, x):
        features = self.base(x)
        # Fix: Ensure features are flattened for linear layer
        if features.dim() > 2:
            features = F.adaptive_avg_pool2d(features, (1, 1))  # Global average pooling
            features = features.view(features.size(0), -1)      # Flatten
        z_img = self.projection_head(features)
        return z_img

class VICE(nn.Module):
    def __init__(self, vision_base_model, imu_input_channels=6, imu_seq_len=20, latent_dim=128):
        super(VICE, self).__init__()
        self.vision_encoder = VisionEncoderWithProjection(vision_base_model, latent_dim)
        self.imu_encoder = IMUEncoder(input_channels=imu_input_channels, seq_len=imu_seq_len, latent_dim=latent_dim)

    def forward(self, img_batch, imu_batch):
        z_img = self.vision_encoder(img_batch)
        z_imu = self.imu_encoder(imu_batch)
        return z_img, z_imu

def info_nce_loss(z_img, z_imu, temperature=0.1):
    batch_size = z_img.size(0)

    z_img = F.normalize(z_img, dim=1)
    z_imu = F.normalize(z_imu, dim=1)

    sim_matrix = torch.matmul(z_img, z_imu.T) / temperature
    labels = torch.arange(batch_size, device=z_img.device)

    loss_img2imu = F.cross_entropy(sim_matrix, labels)
    loss_imu2img = F.cross_entropy(sim_matrix.T, labels)

    loss = (loss_img2imu + loss_imu2img) / 2.0
    return loss

# Custom Dataset
class SynchronizedDataset(Dataset):
    def __init__(self, pairs, img_size=(224, 224), augment=False):
        self.pairs = pairs
        self.img_size = img_size
        self.augment = augment
        self.transform = self.get_transform()

    def get_transform(self):
        from torchvision import transforms
        if self.augment:
            return transforms.Compose([
                transforms.Resize(self.img_size),
                transforms.RandomHorizontalFlip(p=0.5),
                transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
                transforms.ToTensor(),
                transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
            ])
        else:
            return transforms.Compose([
                transforms.Resize(self.img_size),
                transforms.ToTensor(),
                transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
            ])

    def __len__(self):
        return len(self.pairs)

    def __getitem__(self, idx):
        pair = self.pairs[idx]

        # Load image
        img = Image.open(pair['image_path']).convert('RGB')
        img_tensor = self.transform(img)

        # Load IMU data (already preprocessed)
        imu_data = torch.from_numpy(pair['imu_data']).float().T  # Shape: (6, 20)

        return img_tensor, imu_data

# Training function
def train_vice_model(model, train_loader, val_loader, num_epochs=50, lr=1e-4, device='cuda'):
    optimizer = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-4)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=num_epochs)

    train_losses = []
    val_losses = []

    best_val_loss = float('inf')
    patience = 10
    patience_counter = 0

    for epoch in range(num_epochs):
        # Training phase
        model.train()
        train_loss = 0.0
        for batch_idx, (img_batch, imu_batch) in enumerate(train_loader):
            img_batch = img_batch.to(device)
            imu_batch = imu_batch.to(device)

            optimizer.zero_grad()
            z_img, z_imu = model(img_batch, imu_batch)
            loss = info_nce_loss(z_img, z_imu)
            loss.backward()

            # Gradient clipping
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

            optimizer.step()
            train_loss += loss.item()

        avg_train_loss = train_loss / len(train_loader)

        # Validation phase
        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for img_batch, imu_batch in val_loader:
                img_batch = img_batch.to(device)
                imu_batch = imu_batch.to(device)

                z_img, z_imu = model(img_batch, imu_batch)
                loss = info_nce_loss(z_img, z_imu)
                val_loss += loss.item()

        avg_val_loss = val_loss / len(val_loader)

        train_losses.append(avg_train_loss)
        val_losses.append(avg_val_loss)

        print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}')

        # Learning rate scheduling
        scheduler.step()

        # Early stopping
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            patience_counter = 0
            # Save best model
            torch.save(model.state_dict(), 'best_vice_model.pth')
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print(f'Early stopping at epoch {epoch+1}')
                break

    return train_losses, val_losses

# Visualization function
def visualize_embeddings(z_img, z_imu, n_samples=100):
    from sklearn.decomposition import PCA
    import matplotlib.pyplot as plt

    # Convert to numpy and take subset
    z_img_np = z_img.cpu().numpy()[:n_samples]
    z_imu_np = z_imu.cpu().numpy()[:n_samples]

    # Apply PCA for visualization
    pca = PCA(n_components=2)
    z_img_2d = pca.fit_transform(z_img_np)
    z_imu_2d = pca.fit_transform(z_imu_np)

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.scatter(z_img_2d[:, 0], z_img_2d[:, 1], alpha=0.6, label='Images')
    plt.title('Image Embeddings (PCA)')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.scatter(z_imu_2d[:, 0], z_imu_2d[:, 1], alpha=0.6, label='IMU', color='red')
    plt.title('IMU Embeddings (PCA)')
    plt.legend()

    plt.tight_layout()
    plt.show()

```

##Summary of V-ICE Model Development and Training

This section summarizes the architectural improvements, hyperparameter tuning, and training enhancements implemented for the V-ICE model, followed by a discussion of their impact on training and validation loss. It concludes with suggestions for future optimizations.

### Implemented Modifications

1.  **Architectural Improvements:**
    *   **`IMUEncoder`**: Added `nn.Dropout(0.1)` layers after each `nn.ReLU` activation in the convolutional and fully connected layers to reduce overfitting.
    *   **`VisionEncoderWithProjection`**:
        *   Corrected the `base_output_dim` retrieval for the vision backbone, defaulting to `2048` if not explicitly found in the `base_model` object.
        *   Ensured features from the vision backbone are correctly flattened using `F.adaptive_avg_pool2d` and `view` before being passed to the projection head, handling cases where the backbone might output a spatial feature map.
        *   Added `nn.Dropout(0.1)` after the first `nn.ReLU` in the `projection_head` for regularization.
    *   **`VICE` Model**: Integrated the updated `IMUEncoder` and `VisionEncoderWithProjection` with `ResNet50` (pretrained on ImageNet) as the vision backbone, with its classification head removed to serve as a feature extractor.

2.  **Hyperparameter Tuning and Training Enhancements:**
    *   **Optimizer**: Switched from a generic Adam to `torch.optim.AdamW` with a weight decay of `1e-4` for better regularization.
    *   **Learning Rate Scheduler**: Introduced `torch.optim.lr_scheduler.CosineAnnealingLR` with `T_max=num_epochs` to vary the learning rate during training, typically leading to better convergence.

#### 1. `IMUEncoder` Class

*   **Architecture:** The `IMUEncoder` uses a 2-layer 1D Convolutional Neural Network (CNN) followed by an `AdaptiveAvgPool1d` layer, and then two fully connected (linear) layers. This is a common and reasonable approach for processing sequential data like IMU, extracting features using convolutions and then reducing dimensionality for a latent representation.
*   **Layers:**
    *   `conv1` and `conv2`: Both use a `kernel_size=3` and `padding=1`, which means they maintain the sequence length while performing convolutions. This allows for local feature extraction across the IMU time series.
    *   `AdaptiveAvgPool1d(1)`: This layer reduces the temporal dimension of the feature maps to 1, effectively performing global average pooling. This is a standard technique to aggregate features across the sequence before feeding them into a fully connected layer.
    *   `fc1` and `fc2`: These are linear layers that project the pooled convolutional features into the `latent_dim` space. The second `fc2` layer directly outputs the latent representation, which is common for embedding tasks.
*   **Activation Functions:** `ReLU` is used after each convolutional layer and the first fully connected layer. ReLU is a good choice for its computational efficiency and ability to mitigate vanishing gradients.
*   **Dropout:** Dropout layers with a rate of 0.1 are applied after `conv1`, `conv2`, and `fc1`. #### 3. `VICE` (Vision-IMU Contrastive Encoder) Architecture

*   **Integration:** The `VICE` model acts as a wrapper, encapsulating both the `VisionEncoderWithProjection` and `IMUEncoder`. This clear separation of concerns makes the model modular and easier to understand and debug.
*   **Forward Pass:** The `forward` method simply calls the individual encoders for image and IMU batches, returning their respective latent representations (`z_img`, `z_imu`). This is precisely what's needed for contrastive learning, where pairs of embeddings are compared.

#### 4. `info_nce_loss` Implementation

*   **Purpose:** This function implements the InfoNCE loss, a standard objective for contrastive learning, aiming to pull positive pairs (matched image-IMU) closer and push negative pairs apart.
*   **Normalization:** `F.normalize(..., dim=1)` is applied to both `z_img` and `z_imu` embeddings. This is a critical step in contrastive learning, as it projects the embeddings onto a unit hypersphere, making cosine similarity directly equivalent to a dot product and improving training stability.
*   **Similarity Matrix:** `torch.matmul(z_img, z_imu.T) / temperature` computes the similarity between all image embeddings and all IMU embeddings in the batch. The `temperature` parameter scales the logits before the softmax operation. A smaller temperature makes the distribution sharper, forcing the model to learn more discriminative embeddings. This is a crucial hyperparameter for contrastive learning.
*   **Labels and Cross-Entropy:** `labels = torch.arange(batch_size, device=z_img.device)` creates labels where the `i`-th image is considered a positive pair with the `i`-th IMU sample. `F.cross_entropy` is then used, treating the similarity scores as logits. The loss is computed symmetrically (`loss_img2imu` and `loss_imu2img`) to ensure that both modalities are equally optimized to predict their counterparts.

    *   `ToTensor()`: Converts PIL Image to PyTorch tensor (H, W, C) to (C, H, W) and scales pixel values to [0, 1].

```python
import os
import numpy as np
import pickle
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import time
from torchvision import transforms
from torchvision.models import resnet50

# Your model code - MODIFIED
class IMUEncoder(nn.Module):
    def __init__(self, input_channels=6, seq_len=20, latent_dim=256): # latent_dim changed to 256
        super(IMUEncoder, self).__init__()
        self.conv1 = nn.Conv1d(input_channels, 32, kernel_size=3, padding=1)
        self.conv2 = nn.Conv1d(32, 64, kernel_size=3, padding=1)
        self.conv3 = nn.Conv1d(64, 128, kernel_size=3, padding=1) # Added third convolutional layer
        self.pool = nn.AdaptiveAvgPool1d(1)
        self.fc1 = nn.Linear(128, latent_dim) # Input dimension changed from 64 to 128, output to latent_dim
        self.fc2 = nn.Linear(latent_dim, latent_dim) # Input and output dimension changed to latent_dim
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.2) # Dropout rate changed to 0.2

    def forward(self, x):
        x = self.relu(self.conv1(x))
        x = self.dropout(x)
        x = self.relu(self.conv2(x))
        x = self.dropout(x)
        x = self.relu(self.conv3(x)) # Added forward pass for conv3
        x = self.dropout(x)
        x = self.pool(x)
        x = x.view(x.size(0), -1)  # Shape: (batch_size, 128)
        x = self.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x

class VisionEncoderWithProjection(nn.Module):
    def __init__(self, base_model, latent_dim=256): # latent_dim changed to 256
        super(VisionEncoderWithProjection, self).__init__()
        self.base = base_model
        base_output_dim = getattr(base_model, 'num_features', 2048)
        self.projection_head = nn.Sequential(
            nn.Linear(base_output_dim, latent_dim), # Output to latent_dim (256)
            nn.ReLU(),
            nn.Dropout(0.2), # Dropout rate changed to 0.2
            nn.Linear(latent_dim, latent_dim), # Added additional linear layer
            nn.ReLU(),
            nn.Dropout(0.2), # Dropout rate changed to 0.2
            nn.Linear(latent_dim, latent_dim)
        )

    def forward(self, x):
        features = self.base(x)
        if features.dim() > 2:
            features = F.adaptive_avg_pool2d(features, (1, 1))
            features = features.view(features.size(0), -1)
        z_img = self.projection_head(features)
        return z_img

class VICE(nn.Module): # Vision-IMU Contrastive Encoder
    def __init__(self, vision_base_model, imu_input_channels=6, imu_seq_len=20, latent_dim=256): # latent_dim changed to 256
        super(VICE, self).__init__()
        self.vision_encoder = VisionEncoderWithProjection(vision_base_model, latent_dim) # Pass latent_dim
        self.imu_encoder = IMUEncoder(input_channels=imu_input_channels, seq_len=imu_seq_len, latent_dim=latent_dim) # Pass latent_dim

    def forward(self, img_batch, imu_batch):
        z_img = self.vision_encoder(img_batch)
        z_imu = self.imu_encoder(imu_batch)
        return z_img, z_imu

def info_nce_loss(z_img, z_imu, temperature=0.07): # temperature changed to 0.07
    batch_size = z_img.size(0)

    z_img = F.normalize(z_img, dim=1)
    z_imu = F.normalize(z_imu, dim=1)

    sim_matrix = torch.matmul(z_img, z_imu.T) / temperature
    labels = torch.arange(batch_size, device=z_img.device)

    loss_img2imu = F.cross_entropy(sim_matrix, labels)
    loss_imu2img = F.cross_entropy(sim_matrix.T, labels)

    loss = (loss_img2imu + loss_imu2img) / 2.0
    return loss

# Custom Dataset
class SynchronizedDataset(Dataset):
    def __init__(self, pairs, img_size=(224, 224), augment=False):
        self.pairs = pairs
        self.img_size = img_size
        self.augment = augment
        self.transform = self.get_transform()

    def get_transform(self):
        # from torchvision import transforms # Already imported at top
        if self.augment:
            return transforms.Compose([
                transforms.Resize(self.img_size),
                transforms.RandomHorizontalFlip(p=0.5),
                transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
                transforms.ToTensor(),
                transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
            ])
        else:
            return transforms.Compose([
                transforms.Resize(self.img_size),
                transforms.ToTensor(),
                transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
            ])

    def __len__(self):
        return len(self.pairs)

    def __getitem__(self, idx):
        pair = self.pairs[idx]

        img = Image.open(pair['image_path']).convert('RGB')
        img_tensor = self.transform(img)

        imu_data = torch.from_numpy(pair['imu_data']).float().T

        return img_tensor, imu_data

# Training function
def train_vice_model(model, train_loader, val_loader, num_epochs=50, lr=1e-4, device='cuda', start_epoch=0, output_dir='/content/drive/MyDrive/tartanair_v2'):
    optimizer = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-4)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=num_epochs)

    train_losses = []
    val_losses = []

    best_val_loss = float('inf')
    patience = 10
    patience_counter = 0

    # Define paths for saving/loading model and losses
    model_save_path = os.path.join(output_dir, 'best_vice_model.pth')
    losses_save_path = os.path.join(output_dir, 'training_losses.pkl')

    if os.path.exists(model_save_path) and start_epoch > 0:
        print(f"Loading model from {model_save_path} to resume training from epoch {start_epoch}")
        model.load_state_dict(torch.load(model_save_path, map_location=device))

    if os.path.exists(losses_save_path) and start_epoch > 0:
        with open(losses_save_path, 'rb') as f:
            prev_losses = pickle.load(f)
            train_losses = prev_losses['train_losses']
            val_losses = prev_losses['val_losses']
            best_val_loss = min(val_losses) # Update best_val_loss based on loaded history
            print(f"Loaded previous loss history. Last val loss: {val_losses[-1]:.4f}")

    for epoch in range(start_epoch, num_epochs):
        model.train()
        current_train_loss = 0.0
        for batch_idx, (img_batch, imu_batch) in enumerate(train_loader):
            img_batch = img_batch.to(device)
            imu_batch = imu_batch.to(device)

            optimizer.zero_grad()
            z_img, z_imu = model(img_batch, imu_batch)
            loss = info_nce_loss(z_img, z_imu) # Using the updated temperature
            loss.backward()

            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

            optimizer.step()
            current_train_loss += loss.item()

        avg_train_loss = current_train_loss / len(train_loader)

        model.eval()
        current_val_loss = 0.0
        with torch.no_grad():
            for img_batch, imu_batch in val_loader:
                img_batch = img_batch.to(device)
                imu_batch = imu_batch.to(device)

                z_img, z_imu = model(img_batch, imu_batch)
                loss = info_nce_loss(z_img, z_imu) # Using the updated temperature
                current_val_loss += loss.item()

        avg_val_loss = current_val_loss / len(val_loader)

        train_losses.append(avg_train_loss)
        val_losses.append(avg_val_loss)

        print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}')

        scheduler.step()

        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            patience_counter = 0
            torch.save(model.state_dict(), model_save_path)
            print(f"-> Model saved to {model_save_path}: New best validation loss achieved.")
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print(f'Early stopping at epoch {epoch+1}')
                break

        with open(losses_save_path, 'wb') as f:
            pickle.dump({'train_losses': train_losses, 'val_losses': val_losses}, f)

    return train_losses, val_losses

# Visualization function
def visualize_embeddings(z_img, z_imu, n_samples=100):
    from sklearn.decomposition import PCA

    z_img_np = z_img.cpu().numpy()[:n_samples]
    z_imu_np = z_imu.cpu().numpy()[:n_samples]

    pca = PCA(n_components=2)
    z_img_2d = pca.fit_transform(z_img_np)
    z_imu_2d = pca.fit_transform(z_imu_np)

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.scatter(z_img_2d[:, 0], z_img_2d[:, 1], alpha=0.6, label='Images')
    plt.title('Image Embeddings (PCA)')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.scatter(z_imu_2d[:, 0], z_imu_2d[:, 1], alpha=0.6, label='IMU', color='red')
    plt.title('IMU Embeddings (PCA)')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Main execution
if __name__ == "__main__":
    root_path = "/content/drive/MyDrive/tartanair_v2"
    env = "Gascola"
    traj = "easy"
    output_dir = os.path.join(root_path, env, f"Data_{traj}")

    # Ensure the output directory exists in Google Drive
    os.makedirs(output_dir, exist_ok=True)

    with open(os.path.join(output_dir, "pairs_step1.pkl"), "rb") as f:
        pairs = pickle.load(f)

    print(f"Total pairs: {len(pairs)}")

    train_pairs, temp_pairs = train_test_split(pairs, test_size=0.3, random_state=42)
    val_pairs, test_pairs = train_test_split(temp_pairs, test_size=0.5, random_state=42)

    print(f"Train: {len(train_pairs)}, Val: {len(val_pairs)}, Test: {len(test_pairs)}")

    train_dataset = SynchronizedDataset(train_pairs, augment=True)
    val_dataset = SynchronizedDataset(val_pairs, augment=False)

    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2, pin_memory=True) # num_workers reduced to 2
    val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2, pin_memory=True) # num_workers reduced to 2

    vision_model = resnet50(pretrained=True)
    vision_model = nn.Sequential(*list(vision_model.children())[:-1])
    vision_model.num_features = 2048

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = VICE(vision_model, imu_seq_len=20, latent_dim=256).to(device) # latent_dim changed to 256

    print(f"Model initialized on {device}")
    print(f"Total parameters: {sum(p.numel() for p in model.parameters()):,}")

    print("Testing forward pass...")
    dummy_img = torch.randn(2, 3, 224, 224).to(device)
    dummy_imu = torch.randn(2, 6, 20).to(device)
    try:
        z_img, z_imu = model(dummy_img, dummy_imu)
        print(f"Forward pass successful! z_img: {z_img.shape}, z_imu: {z_imu.shape}")
    except Exception as e:
        print(f"Forward pass failed: {e}")

    num_epochs = 100 # Total epochs increased to 100
    start_epoch_resume = 28 # Resuming from epoch 28

    train_losses, val_losses = train_vice_model(model, train_loader, val_loader, num_epochs=num_epochs, lr=1e-4, device=device, start_epoch=start_epoch_resume, output_dir=output_dir)

    plt.figure(figsize=(10, 5))
    plt.plot(range(len(train_losses)), train_losses, label='Training Loss')
    plt.plot(range(len(val_losses)), val_losses, label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Load best model for final evaluation (from Google Drive path)
    model.load_state_dict(torch.load(os.path.join(output_dir, 'best_vice_model.pth'), map_location=device))
    model.eval()

    print("Training completed! Model saved as 'best_vice_model.pth'")
```

#V-ICE Model Architecture (Vision-IMU Contrastive Encoder)

Input:
  - Image Batch (Batch_Size, 3, 224, 224)
  - IMU Batch (Batch_Size, 6, 20)

Components:

1.  Vision Encoder (VisionEncoderWithProjection)
    ├── Base Vision Model (ResNet50, pretrained on ImageNet)
    │   └── Feature Extractor (Outputs 2048 features)
    ├── Global Average Pooling (if needed to flatten features)
    └── Projection Head (nn.Sequential)
        ├── Linear Layer (2048 -> 256)
        ├── ReLU Activation
        ├── Dropout (p=0.3)
        ├── Linear Layer (256 -> 256)        
        ├── ReLU Activation
        ├── Dropout (p=0.3)
        └── Linear Layer (256 -> 256)
    └── Output: Image Embedding (z_img) (Batch_Size, 256)

2.  IMU Encoder (IMUEncoder)
    ├── Conv1d Layer 1 (Input Channels=6 -> 32 features, Kernel=3, Padding=1)
    ├── ReLU Activation
    ├── Dropout (p=0.3)
    ├── Conv1d Layer 2 (32 -> 64 features, Kernel=3, Padding=1)
    ├── ReLU Activation
    ├── Dropout (p=0.3)
    ├── Conv1d Layer 3 (64 -> 128 features, Kernel=3, Padding=1)
    ├── ReLU Activation
    ├── Dropout (p=0.3)
    ├── AdaptiveAvgPool1d (Reduces sequence length to 1)
    ├── Flatten (128 features)
    ├── Linear Layer 1 (128 -> 256)
    ├── ReLU Activation
    ├── Dropout (p=0.3)
    └── Linear Layer 2 (256 -> 256)
    └── Output: IMU Embedding (z_imu) (Batch_Size, 256)

Loss Function:
  - InfoNCE Loss (z_img, z_imu, temperature=0.1)
    ├── Normalization of z_img and z_imu
    ├── Cosine Similarity Matrix Computation
    └── Symmetric Cross-Entropy Calculation

Output:
  - z_img (Image Embeddings)
  - z_imu (IMU Embeddings)

```python
import os
import numpy as np
import pickle
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import time
from torchvision import transforms
from torchvision.models import resnet50

# Your model code - MODIFIED for overfitting fix
class IMUEncoder(nn.Module):
    def __init__(self, input_channels=6, seq_len=20, latent_dim=256):
        super(IMUEncoder, self).__init__()
        self.conv1 = nn.Conv1d(input_channels, 32, kernel_size=3, padding=1)
        self.conv2 = nn.Conv1d(32, 64, kernel_size=3, padding=1)
        self.conv3 = nn.Conv1d(64, 128, kernel_size=3, padding=1)
        self.pool = nn.AdaptiveAvgPool1d(1)
        self.fc1 = nn.Linear(128, latent_dim)
        self.fc2 = nn.Linear(latent_dim, latent_dim)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.3) # Dropout rate changed to 0.3

    def forward(self, x):
        x = self.relu(self.conv1(x))
        x = self.dropout(x)
        x = self.relu(self.conv2(x))
        x = self.dropout(x)
        x = self.relu(self.conv3(x))
        x = self.dropout(x)
        x = self.pool(x)
        x = x.view(x.size(0), -1)
        x = self.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x

class VisionEncoderWithProjection(nn.Module):
    def __init__(self, base_model, latent_dim=256):
        super(VisionEncoderWithProjection, self).__init__()
        self.base = base_model
        base_output_dim = getattr(base_model, 'num_features', 2048)
        self.projection_head = nn.Sequential(
            nn.Linear(base_output_dim, latent_dim),
            nn.ReLU(),
            nn.Dropout(0.3), # Dropout rate changed to 0.3
            nn.Linear(latent_dim, latent_dim),
            nn.ReLU(),
            nn.Dropout(0.3), # Dropout rate changed to 0.3
            nn.Linear(latent_dim, latent_dim)
        )

    def forward(self, x):
        features = self.base(x)
        if features.dim() > 2:
            features = F.adaptive_avg_pool2d(features, (1, 1))
            features = features.view(features.size(0), -1)
        z_img = self.projection_head(features)
        return z_img

class VICE(nn.Module):
    def __init__(self, vision_base_model, imu_input_channels=6, imu_seq_len=20, latent_dim=256):
        super(VICE, self).__init__()
        self.vision_encoder = VisionEncoderWithProjection(vision_base_model, latent_dim)
        self.imu_encoder = IMUEncoder(input_channels=imu_input_channels, seq_len=imu_seq_len, latent_dim=latent_dim)

    def forward(self, img_batch, imu_batch):
        z_img = self.vision_encoder(img_batch)
        z_imu = self.imu_encoder(imu_batch)
        return z_img, z_imu

def info_nce_loss(z_img, z_imu, temperature=0.1): # temperature changed to 0.1
    batch_size = z_img.size(0)

    z_img = F.normalize(z_img, dim=1)
    z_imu = F.normalize(z_imu, dim=1)

    sim_matrix = torch.matmul(z_img, z_imu.T) / temperature
    labels = torch.arange(batch_size, device=z_img.device)

    loss_img2imu = F.cross_entropy(sim_matrix, labels)
    loss_imu2img = F.cross_entropy(sim_matrix.T, labels)

    loss = (loss_img2imu + loss_imu2img) / 2.0
    return loss

# Custom Dataset
class SynchronizedDataset(Dataset):
    def __init__(self, pairs, img_size=(224, 224), augment=False):
        self.pairs = pairs
        self.img_size = img_size
        self.augment = augment
        self.transform = self.get_transform()

    def get_transform(self):
        if self.augment:
            return transforms.Compose([
                transforms.Resize(self.img_size),
                transforms.RandomHorizontalFlip(p=0.5),
                transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
                transforms.ToTensor(),
                transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
            ])
        else:
            return transforms.Compose([
                transforms.Resize(self.img_size),
                transforms.ToTensor(),
                transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
            ])

    def __len__(self):
        return len(self.pairs)

    def __getitem__(self, idx):
        pair = self.pairs[idx]

        img = Image.open(pair['image_path']).convert('RGB')
        img_tensor = self.transform(img)

        imu_data = torch.from_numpy(pair['imu_data']).float().T

        return img_tensor, imu_data

# Training function
def train_vice_model(model, train_loader, val_loader, num_epochs=50, lr=1e-4, device='cuda', start_epoch=0, output_dir='.'):
    optimizer = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-4)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=num_epochs)

    train_losses = []
    val_losses = []

    best_val_loss = float('inf')
    patience = 10
    patience_counter = 0

    model_save_path = os.path.join(output_dir, 'best_vice_model_overfit_fix.pth')
    losses_save_path = os.path.join(output_dir, 'training_losses_overfit_fix.pkl')

    if os.path.exists(model_save_path) and start_epoch > 0:
        print(f"Loading model from {model_save_path} to resume training from epoch {start_epoch}")
        model.load_state_dict(torch.load(model_save_path, map_location=device))

    if os.path.exists(losses_save_path) and start_epoch > 0:
        with open(losses_save_path, 'rb') as f:
            prev_losses = pickle.load(f)
            train_losses = prev_losses['train_losses']
            val_losses = prev_losses['val_losses']
            best_val_loss = min(val_losses) # Update best_val_loss based on loaded history
            print(f"Loaded previous loss history. Last val loss: {val_losses[-1]:.4f}")

    for epoch in range(start_epoch, num_epochs):
        model.train()
        current_train_loss = 0.0
        for batch_idx, (img_batch, imu_batch) in enumerate(train_loader):
            img_batch = img_batch.to(device)
            imu_batch = imu_batch.to(device)

            optimizer.zero_grad()
            z_img, z_imu = model(img_batch, imu_batch)
            loss = info_nce_loss(z_img, z_imu)
            loss.backward()

            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

            optimizer.step()
            current_train_loss += loss.item()

        avg_train_loss = current_train_loss / len(train_loader)

        model.eval()
        current_val_loss = 0.0
        with torch.no_grad():
            for img_batch, imu_batch in val_loader:
                img_batch = img_batch.to(device)
                imu_batch = imu_batch.to(device)

                z_img, z_imu = model(img_batch, imu_batch)
                loss = info_nce_loss(z_img, z_imu)
                current_val_loss += loss.item()

        avg_val_loss = current_val_loss / len(val_loader)

        train_losses.append(avg_train_loss)
        val_losses.append(avg_val_loss)

        print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}')

        scheduler.step()

        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            patience_counter = 0
            torch.save(model.state_dict(), model_save_path)
            print(f"-> Model saved to {model_save_path}: New best validation loss achieved.")
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print(f'Early stopping at epoch {epoch+1}')
                break

        with open(losses_save_path, 'wb') as f:
            pickle.dump({'train_losses': train_losses, 'val_losses': val_losses}, f)

    return train_losses, val_losses

# Visualization function
def visualize_embeddings(z_img, z_imu, n_samples=100):
    from sklearn.decomposition import PCA

    z_img_np = z_img.cpu().numpy()[:n_samples]
    z_imu_np = z_imu.cpu().numpy()[:n_samples]

    pca = PCA(n_components=2)
    z_img_2d = pca.fit_transform(z_img_np)
    z_imu_2d = pca.fit_transform(z_imu_np)

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.scatter(z_img_2d[:, 0], z_img_2d[:, 1], alpha=0.6, label='Images')
    plt.title('Image Embeddings (PCA)')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.scatter(z_imu_2d[:, 0], z_imu_2d[:, 1], alpha=0.6, label='IMU', color='red')
    plt.title('IMU Embeddings (PCA)')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Main execution
if __name__ == "__main__":
    root_path = "/content/drive/MyDrive/tartanair_v2"
    env = "Gascola"
    traj = "easy"
    output_dir = os.path.join(root_path, env, f"Data_{traj}")

    os.makedirs(output_dir, exist_ok=True)

    with open(os.path.join(output_dir, "pairs_step1.pkl"), "rb") as f:
        pairs = pickle.load(f)

    print(f"Total pairs: {len(pairs)}")

    train_pairs, temp_pairs = train_test_split(pairs, test_size=0.3, random_state=42)
    val_pairs, test_pairs = train_test_split(temp_pairs, test_size=0.5, random_state=42)

    print(f"Train: {len(train_pairs)}, Val: {len(val_pairs)}, Test: {len(test_pairs)}")

    train_dataset = SynchronizedDataset(train_pairs, augment=True)
    val_dataset = SynchronizedDataset(val_pairs, augment=False)

    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2, pin_memory=True)
    val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2, pin_memory=True)

    vision_model = resnet50(pretrained=True)
    vision_model = nn.Sequential(*list(vision_model.children())[:-1])
    vision_model.num_features = 2048

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = VICE(vision_model, imu_seq_len=20, latent_dim=256).to(device)

    print(f"Model initialized on {device}")
    print(f"Total parameters: {sum(p.numel() for p in model.parameters()):,}")

    print("Testing forward pass...")
    dummy_img = torch.randn(2, 3, 224, 224).to(device)
    dummy_imu = torch.randn(2, 6, 20).to(device)
    try:
        z_img, z_imu = model(dummy_img, dummy_imu)
        print(f"Forward pass successful! z_img: {z_img.shape}, z_imu: {z_imu.shape}")
    except Exception as e:
        print(f"Forward pass failed: {e}")

    num_epochs = 100
    start_epoch_resume = 41 # Starting from 0 for a fresh run with new hyperparameters

    train_losses, val_losses = train_vice_model(model, train_loader, val_loader, num_epochs=num_epochs, lr=1e-4, device=device, start_epoch=start_epoch_resume, output_dir=output_dir)

    plt.figure(figsize=(10, 5))
    plt.plot(range(len(train_losses)), train_losses, label='Training Loss')
    plt.plot(range(len(val_losses)), val_losses, label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()

    model.load_state_dict(torch.load(os.path.join(output_dir, 'best_vice_model_overfit_fix.pth'), map_location=device))
    model.eval()

    print("Training completed! Model saved as 'best_vice_model_overfit_fix.pth'")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import pickle
import os
from PIL import Image
import torchvision.transforms as transforms

# Assuming the model definition code (IMUEncoder, VisionEncoderWithProjection, VICE, info_nce_loss)
# and the SynchronizedDataset class are available from your training script.

# --- 1. Load the Trained Model ---
def load_trained_model(model_path, vision_base_model, device, imu_seq_len=20, latent_dim=256):
    model = VICE(vision_base_model, imu_seq_len=imu_seq_len, latent_dim=latent_dim).to(device)
    model.load_state_dict(torch.load(model_path, map_location=device))
    model.eval()
    return model

# --- 2. Load Test Data (from original trajectory) ---
def load_test_data(test_pairs_path, img_size=(224, 224)):
    with open(test_pairs_path, "rb") as f:
        pairs = pickle.load(f)

    test_dataset = SynchronizedDataset(pairs, img_size=img_size, augment=False)
    test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False, num_workers=2, pin_memory=True)
    return test_loader, pairs

# --- 3. Helper Function to Get Embeddings ---
@torch.no_grad()
def get_embeddings(model, data_loader, device):
    model.eval()
    all_z_img = []
    all_z_imu = []
    all_img_paths = [] # Store image paths for potential analysis
    all_imu_data = []  # Store raw IMU data for potential analysis

    for img_batch, imu_batch in data_loader:
        img_batch = img_batch.to(device)
        imu_batch = imu_batch.to(device)

        z_img, z_imu = model(img_batch, imu_batch)

        all_z_img.append(z_img.cpu())
        all_z_imu.append(z_imu.cpu())

    all_z_img = torch.cat(all_z_img, dim=0)
    all_z_imu = torch.cat(all_z_imu, dim=0)
    return all_z_img, all_z_imu

# --- 4. Test Functions ---

def test_similarity_metrics(z_img, z_imu, top_k=1):
    """Calculate similarity-based metrics like retrieval accuracy."""
    # Normalize embeddings
    z_img_norm = F.normalize(z_img, dim=1)
    z_imu_norm = F.normalize(z_imu, dim=1)

    # Compute similarity matrix (cosine similarity)
    sim_matrix = torch.matmul(z_img_norm, z_imu_norm.T) # Shape: (N, N)

    # Diagonal elements are the correct matches (image i matches IMU i)
    labels = torch.arange(z_img.size(0)).to(z_img.device)

    # --- Retrieval Accuracy (Top-K) ---
    _, indices_img2imu = torch.topk(sim_matrix, k=top_k, dim=1) # Shape: (N, K)
    correct_img2imu = torch.any(indices_img2imu == labels.unsqueeze(1), dim=1)
    accuracy_img2imu = correct_img2imu.float().mean().item()

    # For each IMU embedding, find the closest image embedding and check if it's the correct one.
    _, indices_imu2img = torch.topk(sim_matrix.T, k=top_k, dim=1) # Shape: (N, K)
    correct_imu2img = torch.any(indices_imu2img == labels.unsqueeze(1), dim=1)
    accuracy_imu2img = correct_imu2img.float().mean().item()

    mean_accuracy = (accuracy_img2imu + accuracy_imu2img) / 2.0

    # --- Average Similarity for Positive Pairs (diagonal) vs Negative Pairs ---
    positive_similarities = torch.diag(sim_matrix)
    mean_pos_sim = positive_similarities.mean().item()

    # Calculate mean similarity for negative pairs (off-diagonal)
    mask_off_diag = ~torch.eye(z_img.size(0), dtype=torch.bool)
    negative_similarities = sim_matrix[mask_off_diag]
    mean_neg_sim = negative_similarities.mean().item()

    return {
        'retrieval_acc_img2imu': accuracy_img2imu,
        'retrieval_acc_imu2img': accuracy_imu2img,
        'mean_retrieval_acc': mean_accuracy,
        'mean_pos_sim': mean_pos_sim,
        'mean_neg_sim': mean_neg_sim,
        'sim_diff': mean_pos_sim - mean_neg_sim # Difference is a key metric
    }

def visualize_embeddings_2d(z_img, z_imu, method='pca', n_samples=500):
    """Visualize embeddings using PCA or t-SNE."""
    # Limit samples for visualization speed if necessary
    if z_img.size(0) > n_samples:
        indices = np.random.choice(z_img.size(0), n_samples, replace=False)
        z_img = z_img[indices]
        z_imu = z_imu[indices]

    z_img_np = z_img.cpu().numpy()
    z_imu_np = z_imu.cpu().numpy()

    if method == 'pca':
        reducer = PCA(n_components=2)
    elif method == 'tsne':
        reducer = TSNE(n_components=2, random_state=42, perplexity=min(30, len(z_img_np)-1)) # Perplexity must be < n_samples
    else:
        raise ValueError("Method must be 'pca' or 'tsne'")

    # Combine embeddings for joint visualization
    combined_embeddings = np.concatenate([z_img_np, z_imu_np], axis=0)
    combined_2d = reducer.fit_transform(combined_embeddings)

    z_img_2d = combined_2d[:len(z_img_np)]
    z_imu_2d = combined_2d[len(z_img_np):]

    plt.figure(figsize=(10, 8))
    plt.scatter(z_img_2d[:, 0], z_img_2d[:, 1], alpha=0.6, label='Image Embeddings', s=20)
    plt.scatter(z_imu_2d[:, 0], z_imu_2d[:, 1], alpha=0.6, label='IMU Embeddings', s=20, color='red')

    plt.title(f'{method.upper()} Visualization of Embeddings')
    plt.xlabel(f'{method.upper()} Component 1')
    plt.ylabel(f'{method.upper()} Component 2')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

# --- Main Testing Execution ---
if __name__ == "__main__":
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    output_dir = "/content/drive/MyDrive/tartanair_v2/Gascola/Data_easy" # Adjust path if needed
    model_path = os.path.join(output_dir, 'best_vice_model_overfit_fix.pth')

    # --- PART FOR SAME TRAJECTORY (ORIGINAL TEST SET) ---
    # This part loads the test data from the *same trajectory* used for training/validation splits
    test_pairs_path = os.path.join(output_dir, "pairs_step1.pkl") # This file contains the original data splits

    # Load the trained model
    vision_model = resnet50(pretrained=True)
    vision_model = nn.Sequential(*list(vision_model.children())[:-1])
    vision_model.num_features = 2048
    model = load_trained_model(model_path, vision_model, device)

    # Load test data (from the SAME trajectory as training/validation)
    test_loader, original_pairs = load_test_data(test_pairs_path)
    print(f"Loaded {len(test_loader.dataset)} test pairs from the SAME trajectory.")

    # Get embeddings
    print("Generating embeddings for SAME trajectory test set...")
    z_img_test, z_imu_test = get_embeddings(model, test_loader, device)
    print(f"Got {z_img_test.shape[0]} image and {z_imu_test.shape[0]} IMU embeddings.")

    # Run tests on SAME trajectory data
    print("\n--- Running Similarity Tests (SAME TRAJECTORY) ---")
    metrics_same_traj = test_similarity_metrics(z_img_test, z_imu_test, top_k=1)
    print(f"Same Traj - Mean Retrieval Accuracy: {metrics_same_traj['mean_retrieval_acc']:.4f}")
    print(f"Same Traj - Image -> IMU Retrieval Accuracy: {metrics_same_traj['retrieval_acc_img2imu']:.4f}")
    print(f"Same Traj - IMU -> Image Retrieval Accuracy: {metrics_same_traj['retrieval_acc_imu2img']:.4f}")
    print(f"Same Traj - Mean Similarity (Positive Pairs): {metrics_same_traj['mean_pos_sim']:.4f}")
    print(f"Same Traj - Mean Similarity (Negative Pairs): {metrics_same_traj['mean_neg_sim']:.4f}")
    print(f"Same Traj - Difference (Pos - Neg): {metrics_same_traj['sim_diff']:.4f}")

    print("\n--- Visualizing Embeddings (SAME TRAJECTORY - PCA) ---")
    visualize_embeddings_2d(z_img_test, z_imu_test, method='pca', n_samples=500)

    # # --- PART FOR NEW TRAJECTORY ---
    # # You need to prepare the new trajectory data first using your data loading/sync code.
    # # Save the synchronized pairs from the new trajectory to a new .pkl file.
    # # Then update the path below.
    # new_test_pairs_path = "/content/drive/MyDrive/tartanair_v2/NewEnv/Data_hard/pairs_step1.pkl" # UPDATE THIS PATH

    # if os.path.exists(new_test_pairs_path):
    #     print("\n--- Testing on NEW Trajectory ---")
    #     # Load test data (from a NEW, unseen trajectory)
    #     new_test_loader, _ = load_test_data(new_test_pairs_path)
    #     print(f"Loaded {len(new_test_loader.dataset)} test pairs from the NEW trajectory.")

    #     # Get embeddings for NEW trajectory
    #     print("Generating embeddings for NEW trajectory test set...")
    #     z_img_new, z_imu_new = get_embeddings(model, new_test_loader, device)
    #     print(f"Got {z_img_new.shape[0]} image and {z_imu_new.shape[0]} IMU embeddings.")

    #     # Run tests on NEW trajectory data
    #     new_metrics = test_similarity_metrics(z_img_new, z_imu_new, top_k=1)
    #     print(f"New Traj - Mean Retrieval Accuracy: {new_metrics['mean_retrieval_acc']:.4f}")
    #     print(f"New Traj - Difference (Pos - Neg): {new_metrics['sim_diff']:.4f}")
    # else:
    #     print("\n--- New trajectory data file not found at the specified path. Skipping NEW TRAJECTORY test. ---")
    #     print(f"   Please update 'new_test_pairs_path' to point to your new trajectory's .pkl file.")
    #     print(f"   Path checked: {new_test_pairs_path}")

    # --- Optional: Test with Outliers (Fake Pairs) - Applies to any loaded embeddings ---
    # This test can be run on either the same or new trajectory embeddings.
    # It shuffles the IMU embeddings relative to the images to create mismatched pairs.
    print("\n--- Testing with Fake Pairs (Outliers) on SAME TRAJECTORY Embeddings ---")
    # Shuffle the IMU embeddings along the batch dimension to create mismatched pairs
    shuffled_indices = torch.randperm(z_imu_test.size(0))
    z_imu_shuffled = z_imu_test[shuffled_indices]

    # Calculate similarity metrics between real img and shuffled imu
    fake_metrics = test_similarity_metrics(z_img_test, z_imu_shuffled, top_k=1)
    print(f"Fake Pairs (Same Traj) - Mean Retrieval Accuracy: {fake_metrics['mean_retrieval_acc']:.4f}")
    print(f"Fake Pairs (Same Traj) - Difference (Pos - Neg): {fake_metrics['sim_diff']:.4f}")
    print("Note: Fake pair accuracy should be near random (e.g., 1/N for N samples).")
    print("      The 'sim_diff' for fake pairs should be much lower than for real pairs.")
```
